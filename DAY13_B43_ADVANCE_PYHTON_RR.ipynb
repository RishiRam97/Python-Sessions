## Pie chart making using matplotlib and the bar graphs.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Basic approach:

```python
y = np.array([35,25,25,15])
```

```python
plt.pie(y)
plt.show()
```

#### Req: to add the labels

```python
mylabels = ['Apples','Bananas','Cherries','Dates']
```

```python
plt.pie(y, labels= mylabels)
plt.show()
```

#### How to control the angels on the pie chart....default it will start at 0.

```python
plt.pie(y, labels= mylabels, startangle=90)
plt.show()
```

#### EXPLODE

```python
myexplode = [0.2,0,0,0]
```

```python
plt.pie(y, labels= mylabels, explode = myexplode)
plt.show()
```

```python
plt.pie(y, labels= mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
plt.pie(y, labels= mylabels, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```

### Creating bars with matplotlib:

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])         #VERTICAL BAR
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

### CUSTOM COLOR

```python
plt.bar(x,y, color = 'red')
plt.show()
```

```python
plt.bar(x,y, color = 'hotpink')
plt.show()
```

### WIDTH CONTROL

```python
plt.bar(x,y,width = 0.1)
plt.show()
```
